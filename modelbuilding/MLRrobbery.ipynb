```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("robbery_data.csv")
```

```python
df=df.drop(['District Name'],axis=1)
df=df.drop([17])
```

```python
y=df.pop("No. of Robeery Cases(2019)")
x=df
```

```python
import statsmodels.api as sm
model = sm.OLS(y, x).fit()
model.summary()
```
