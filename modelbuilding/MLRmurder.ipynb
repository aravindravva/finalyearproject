```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("murder_data.csv")
```

```python
df
```

# data cleaning 

```python
df=df.drop(['District Name'],axis=1)
```

```python
df=df.drop([17])
```

```python
y=df.pop('Number of murders(2019)')
x=df
```

# training

```python
import statsmodels.api as sm
model = sm.OLS(y, x).fit()
model.summary()
```

```python
df.columns
```

```python
import pickle
pickle.dump(model,open("MLRmurderprediction.pickle","wb"))
```

```python
modelfile="MLRmurderprediction.pickle"
model=pickle.load(open(modelfile,"rb"))
```
